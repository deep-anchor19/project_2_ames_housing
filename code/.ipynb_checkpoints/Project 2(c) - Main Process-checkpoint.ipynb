## Data Modelling

### Model Prep: Create our features matrix (`X`) and target vector (`y`)

```python
X = train_dummified.drop(columns = ['id','saleprice'])
print(X.shape)
X.head()
```

```python
y = train_dummified['saleprice']
y.head()
```

### Model Prep: Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5, test_size=0.35)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Model Prep: Scaling

```python
ss = StandardScaler()
Z_train = ss.fit_transform(X_train)
Z_test = ss.transform(X_test)
```

### Model Prep: Instantiate Our Models

We'll be evaluating `LinearRegression`, `LassoCV` and `RidgeCV`, so we'll need to create instances of all three models.

#### LinearRegression

```python
# instantiate linear regression model
lr = LinearRegression()

# fit data into the LR model
lr.fit(Z_train, y_train)
```

```python
# iterate thru 3-20 folds to check which no. of folds gives the highest score
for folds in range(3,21):
    print('-----------------------------------------------\n')
    print('K: ', folds)
    
    # perform cross-validation
    scores = cross_val_score(lr, Z_train, y_train, cv=folds)
    print('Cross-Validated Scores: ', scores)
    print('Mean CV R2: ', np.mean(scores))
    print('Standard Deviation CV R2: ', np.std(scores))
    
    # perform cross-validated predictions
    predictions = cross_val_predict(lr, Z_train, y_train, cv=folds)
    r2 = r2_score(y_train, predictions)
    print('Cross-Predicted R2: ', r2)
    
    # display the RMSE for cross-validated predictions
    print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
# get cross validation score of model on train
# use k-fold of 9 as its cross-predicted R2 is the highest (least negative), and RMSE is the lowest
lr_scores = cross_val_score(lr, Z_train, y_train, cv=9)

# print cross-validation score of each fold on train set
print('LR Scores: ', lr_scores)

# print the mean of cross-validation scores on train set
print('Mean LR Score: ', lr_scores.mean())

# print the cross prediction score on train set
predictions = cross_val_predict(lr, Z_train, y_train, cv=9)

# print the mean of the RMSE scores on train set
print('Mean RMSE Score: ', np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
# get cross validation score of model on test
# use k-fold of 9 as its cross-predicted R2 is the highest (least negative), and RMSE is the lowest
lr_scores = cross_val_score(lr, Z_test, y_test, cv=9)

# print cross-validation score of each fold on test set
print('LR Scores: ', lr_scores)

# print the mean of cross-validation scores on test set
print('Mean LR Score: ', lr_scores.mean())

# print the cross prediction score on test set
predictions = cross_val_predict(lr, Z_test, y_test, cv=9)

# print the mean of the RMSE scores on test set
print('Mean RMSE Score: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# print optimal coefficients for Linear Regression
lr.coef_
```

#### RidgeCV

```python
# find optimal value for ridge regression alpha
# np.logspace generates 200 values equally between 0 and 5, then converts them to alphas between 10**0 and 10**5 
ridge_alphas = np.logspace(0, 5, 200)

# cross-validate over our list of ridge aplhas
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=6)

# fit Z_train and y_train into ridge model
optimal_ridge.fit(Z_train, y_train)

# print optimal ridge alpha
print('Optimal Ridge Alpha: ', optimal_ridge.alpha_)
```

```python
# instantiate ridge model with optimal alpha
ridge = Ridge(alpha=optimal_ridge.alpha_)
```

```python
# iterate thru 3-20 folds to check which no. of folds gives the highest score
for folds in range(3,21):
    print('-----------------------------------------------\n')
    print('K: ', folds)
    
    # perform cross-validation
    scores = cross_val_score(ridge, Z_train, y_train, cv=folds)
    print('Cross-Validated Scores: ', scores)
    print('Mean CV R2: ', np.mean(scores))
    print('Standard Deviation CV R2: ', np.std(scores))
    
    # perform cross-validated predictions
    predictions = cross_val_predict(ridge, Z_train, y_train, cv=folds)
    r2 = r2_score(y_train, predictions)
    print('Cross-Predicted R2: ', r2)
    
    # display the RMSE for cross-validated predictions
    print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
# get cross validation score of model on train set

# use k-fold of 3 as its cross-predicted R2 is the highest (least negative), and RMSE is the lowest
ridge_scores = cross_val_score(ridge, Z_train, y_train, cv=3)

# print cross-validation score of each fold on train set
print('Ridge Scores: ', ridge_scores)

# print the mean of cross-validation scores on train set
print('Mean Ridge Score: ', np.mean(ridge_scores))

# print the cross prediction score on train set
predictions = cross_val_predict(ridge, Z_train, y_train, cv=3)

# print RMSE for cross-validated predictions
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
# get cross validation score of model on test set

# use k-fold of 3 as its cross-predicted R2 is the highest (least negative), and RMSE is the lowest
ridge_scores = cross_val_score(ridge, Z_test, y_test, cv=3)

# print cross-validation score of each fold on train set
print('Ridge Scores: ', ridge_scores)

# print the mean of cross-validation scores on train set
print('Mean Ridge Score: ', np.mean(ridge_scores))

# print the cross prediction score on train set
predictions = cross_val_predict(ridge, Z_test, y_test, cv=3)

# print RMSE for cross-validated predictions
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# print optimal coefficients for Ridge
optimal_ridge.coef_
```

#### LassoCV

```python
# find optimal value for regression alpha
optimal_lasso = LassoCV(n_alphas=500, cv=6, verbose=1)

# fit Z_train and y_train into lasso model
optimal_lasso.fit(Z_train, y_train)

# print the optimal lasso alpha
print('Optimal Lasso Alpha: ', optimal_lasso.alpha_)
```

```python
# instantiate lasso model with optimal alpha
lasso = Lasso(alpha=optimal_lasso.alpha_)
```

```python
# iterate thru 3-20 folds to check which no. of folds gives the highest score
for folds in range(3,21):
    print('-----------------------------------------------\n')
    print('K: ', folds)
    
    # perform cross-validation
    scores = cross_val_score(lasso, Z_train, y_train, cv=folds)
    print('Cross-Validated Scores: ', scores)
    print('Mean CV R2: ', np.mean(scores))
    print('Standard Deviation CV R2: ', np.std(scores))
    
    # perform cross-validated predictions
    predictions = cross_val_predict(lasso, Z_train, y_train, cv=folds)
    r2 = r2_score(y_train, predictions)
    print('Cross-Predicted R2: ', r2)
    
    # display the RMSE for cross-validated predictions
    print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
# get cross validation score of model on train set

# use k-fold of 3 as its cross-predicted R2 is the highest (least negative), and RMSE is the lowest
lasso_scores = cross_val_score(lasso, Z_train, y_train, cv=3)

# print cross-validation score of each fold on train set
print('Lasso Scores: ', lasso_scores)

# print the mean of cross-validation scores on train set
print('Mean Lasso Score: ', np.mean(lasso_scores))

# print the cross prediction score on train set
predictions = cross_val_predict(lasso, Z_train, y_train, cv=3)

# print RMSE for cross-validated predictions
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
# get cross validation score of model on test set

# use k-fold of 3 as its cross-predicted R2 is the highest (least negative), and RMSE is the lowest
lasso_scores = cross_val_score(lasso, Z_test, y_test, cv=3)

# print cross-validation score of each fold on train set
print('Lasso Scores: ', lasso_scores)

# print the mean of cross-validation scores on train set
print('Mean Lasso Score: ', np.mean(lasso_scores))

# print the cross prediction score on train set
predictions = cross_val_predict(lasso, Z_test, y_test, cv=3)

# print RMSE for cross-validated predictions
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# print optimal coefficients for Lasso
optimal_lasso.coef_
```

### Saving Coefficients to Dataframe

```python
# sorting coefficient of lasso coefficients to show the features that affects the model most severely
lasso_col_coef = list(zip(X.columns, abs(optimal_lasso.coef_), optimal_lasso.coef_))
lasso_coef_df = pd.DataFrame(lasso_col_coef)
lasso_coef_df = lasso_coef_df.rename(columns={0: 'feature',
                                              1: 'coefficient_absolute',
                                              2: 'coefficient'})
lasso_coef_df = lasso_coef_df.sort_values(by=['coefficient_absolute'], ascending=False)
```

```python
# filtering top 10 positive and top 10 negative lasso coefficients
pos10_coef = lasso_coef_df.sort_values(by=['coefficient'], ascending=False).head(10)
neg10_coef = lasso_coef_df.sort_values(by=['coefficient'], ascending=True).head(10)
pos_neg_coef = pd.concat([pos10_coef, neg10_coef])
print(pos_neg_coef)
```

```python
# plot top 10 positive and top 10 negative coefficients
pos_neg_coef_ax = pd.Series(list(pos_neg_coef['coefficient']), index=pos_neg_coef['feature']).plot(kind='bar', figsize=(25, 10));
pos_neg_coef_ax.set_xticklabels(labels=pos_neg_coef['feature'], rotation=45, ha='right', fontsize=14);
pos_neg_coef_ax.set_xlabel('Features', fontsize=20);
pos_neg_coef_ax.set_ylabel('Coefficient Value', fontsize=20);
pos_neg_coef_ax.set_title('Top 10 Positive and Top 10 Negative Coefficients', pad=10.0, fontsize=24);
pos_neg_coef_ax.axhline(1.25, color='black');
plt.show();
```

## Conclusion and Recommendations

#### Conclusion:

Neighborhood:
- Focus on marketing the houses in Northridge Heights and Northridge.


Quality of House:
- Advice clients with below average housing quality to refurbish their home.


Fireplaces:
- Focus on marketing houses with at least 1 fireplace and encourage clients to build a fireplace if they do not have one.


Housing Exteriors:
- Avoid houses with hardboard or stucco exteriors as they are prone to rot and mold especially with the above average snowfall in Ames.


BIGGER is NOT always BETTER:
- A bigger house may have a higher potential sale price, but that also means that it could have a lower chance of being sold due to a lower number of potential buyers who can afford it


#### Recommendations for the Future:

Total No. of Houses in Neighborhoods
- This would allow us to evaluate the percentage % of houses that are sold across all neighborhoods.


Demographic of Neighborhood and Buyers
- Age range, Marital Status, Family size, No. of kids could play a part in helping realtors recommend the right houses to the right buyers.


Sale price in real value to take into account of inflation
- Rate of inflation would allow us to “reflect” the past sale prices based on today’s economy.
